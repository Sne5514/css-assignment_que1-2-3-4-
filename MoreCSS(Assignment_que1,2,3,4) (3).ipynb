3. Create a navigation bar similar to question 1. For each navigation item, use a different border style.

  -  

***HTML***



```
<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8">
  <meta name="viewport" content="width=device-width, initial-scale=1.0">
  <title>Styled Navigation Bar</title>
  <link rel="stylesheet" href="assignment3.css" >
</head>
<body>
    <nav>
        <ul class="navbar">
            <li class="nav-item solid"><a href="#">Home</a></li>
            <li class="nav-item dashed"><a href="#">About Us</a></li>
            <li class="nav-item dotted"><a href="#">Contact</a></li>
            <li class="nav-item double"><a href="#">Blog</a></li>
            <li class="nav-item groove"><a href="#">Login</a></li>
        </ul>
    </nav>
</body>
</html>
```


**CSS**



```
body {
    font-family: Arial, sans-serif;
    margin: 0;
    padding: 0;
    background-color: #f4f4f4;
}

nav {
    background-color: #333;
    padding: 10px 0;
}

.navbar {
    list-style: none;
    display: flex;
    justify-content: center;
    padding: 0;
}

.nav-item {
    margin: 0 15px;
    padding: 10px 20px;
}

.nav-item a {
    text-decoration: none;
    color: white;
    font-size: 18px;
    padding: 10px 15px;
    display: inline-block;
}

.solid {
    border: 3px solid white;
}

.dashed {
    border: 3px dashed white;
}

.dotted {
    border: 3px dotted white;
}

.double {
    border: 3px double white;
}

.groove {
    border: 3px groove white;
}

/* Hover effect */
.nav-item:hover {
    background-color: #555;
}

```

4. Create a simple webpage with an image and make the image circular using border-radius.


 -


```
<!DOCTYPE html>
<html lang="en">
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Image Circle</title>
    <style>
        body {
            text-align: center;
            font-family: Arial, sans-serif;
            margin: 50px;
        }
        .circular-image {
            width: 200px;  /* Adjust size as needed */
            height: 200px; /* Keep width and height equal for a perfect circle */
            border-radius: 50%;
            object-fit: cover; /* Ensures the image fills the container without distortion */
            border: 5px solid black; /* Optional: adds a border */
        }
    </style>
</head>
<body>

    <h2>Circle Image Example</h2>
    

</body>
</html>
```
